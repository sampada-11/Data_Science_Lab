```python
#multiplication table
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):  # Table from 1 to 10
        print(f"{number} x {i} = {number * i}")

# Example usage:
num = 7
display_multiplication_table(num)

print("\n") # Add a newline for better readability between examples

num = 12
display_multiplication_table(num)
```

```python
#palindrome
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the original string with its reverse
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage:
num1 = 121
print(is_palindrome(num1))

num2 = 123
print(is_palindrome(num2))

num3 = 1001
print(is_palindrome(num3))
```

```python
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
num1 = 10
print(check_odd_even(num1))

num2 = 7
print(check_odd_even(num2))

num3 = 0
print(check_odd_even(num3))
```

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Example usage:
number = 5
print(f"The factorial of {number} is {factorial(number)}")

number = 0
print(f"The factorial of {number} is {factorial(number)}")

number = -3
print(f"The factorial of {number} is {factorial(number)}")
```

# Task
Develop an expert system with a knowledge base defined using Python data structures, an inference engine, and a text-based user interface, then test the system and summarize its capabilities.

## Define Knowledge Base

### Subtask:
Define a simple knowledge base for a specific domain (e.g., animal identification based on characteristics) using Python data structures like dictionaries or lists of rules.


**Reasoning**:
To define the knowledge base as requested, I will create a Python code block. This block will contain a list of dictionaries, where each dictionary represents a rule for animal identification with 'conditions' and 'conclusion' keys, as specified in the instructions. I will include at least 5-7 such rules.



```python
knowledge_base = [
    {
        'conditions': ['has fur', 'barks', 'is domesticated'],
        'conclusion': 'Dog'
    },
    {
        'conditions': ['has fur', 'meows', 'is domesticated'],
        'conclusion': 'Cat'
    },
    {
        'conditions': ['has feathers', 'can fly', 'lays eggs'],
        'conclusion': 'Bird'
    },
    {
        'conditions': ['has scales', 'lays eggs', 'no legs'],
        'conclusion': 'Snake'
    },
    {
        'conditions': ['lives in water', 'has fins', 'has gills'],
        'conclusion': 'Fish'
    },
    {
        'conditions': ['has rough skin', 'has four legs', 'lives in water and land'],
        'conclusion': 'Frog'
    },
    {
        'conditions': ['has long trunk', 'has large ears', 'is grey'],
        'conclusion': 'Elephant'
    }
]

print("Knowledge Base defined with the following rules:")
for i, rule in enumerate(knowledge_base):
    print(f"Rule {i+1}: If characteristics are {rule['conditions']}, then it's a {rule['conclusion']}")
```

## Implement Inference Engine

### Subtask:
Develop a Python function that acts as the inference engine. This engine will take user inputs (characteristics) and apply the rules from the knowledge base to derive a decision or classification.


**Reasoning**:
To implement the inference engine, I will define a Python function `infer_animal` that iterates through the `knowledge_base` and checks if all conditions of a rule are present in the user's provided characteristics to infer an animal.



```python
def infer_animal(user_characteristics, knowledge_base):
    for rule in knowledge_base:
        # Check if all conditions of the rule are present in user_characteristics
        if all(condition in user_characteristics for condition in rule['conditions']):
            return rule['conclusion']
    return "Unknown animal based on provided characteristics"

# Demonstrate usage with examples
print("--- Inference Engine Examples ---")

# Example 1: Characteristics for a Dog
user_chars_dog = ['has fur', 'barks', 'is domesticated', 'loyal']
print(f"Characteristics: {user_chars_dog}")
print(f"Inferred Animal: {infer_animal(user_chars_dog, knowledge_base)}\n")

# Example 2: Characteristics for a Fish
user_chars_fish = ['lives in water', 'has fins', 'has gills']
print(f"Characteristics: {user_chars_fish}")
print(f"Inferred Animal: {infer_animal(user_chars_fish, knowledge_base)}\n")

# Example 3: Characteristics for an unknown animal
user_chars_unknown = ['has scales', 'can fly', 'likes bananas']
print(f"Characteristics: {user_chars_unknown}")
print(f"Inferred Animal: {infer_animal(user_chars_unknown, knowledge_base)}\n")

# Example 4: Characteristics for a Frog
user_chars_frog = ['has rough skin', 'has four legs', 'lives in water and land', 'jumps']
print(f"Characteristics: {user_chars_frog}")
print(f"Inferred Animal: {infer_animal(user_chars_frog, knowledge_base)}\n")
```

## Create User Interface

### Subtask:
Build a basic text-based user interface that allows the user to input information and receive the expert system's decision or recommendation.


**Reasoning**:
To build the text-based user interface, I will write a Python code block that prompts the user for characteristics, processes the input, uses the previously defined `infer_animal` function, and displays the result. I will also include a loop and a quit option for better usability.



```python
print("\n--- Expert System User Interface ---")
print("Enter animal characteristics separated by commas (e.g., 'has fur, barks, is domesticated').")
print("Type 'quit' to exit.\n")

while True:
    user_input = input("Enter characteristics: ").strip().lower()

    if user_input == 'quit':
        print("Exiting expert system. Goodbye!")
        break

    # Convert the user's input string into a list of individual characteristics
    user_characteristics = [char.strip() for char in user_input.split(',') if char.strip()]

    if not user_characteristics:
        print("No characteristics entered. Please try again or type 'quit' to exit.\n")
        continue

    # Call the infer_animal function and print the result
    inferred_animal = infer_animal(user_characteristics, knowledge_base)
    print(f"Based on the characteristics, the animal is: {inferred_animal}\n")
```
