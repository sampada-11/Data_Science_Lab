```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Iris.csv')

# --- 1. Describe the dataset ---
print("--- Dataset Description ---")
print("\nPurpose:")
print("The Iris dataset is a classic and very popular dataset for machine learning and statistics, often used for classification tasks. It contains measurements of various parts of Iris flowers.")
print("\nKey Features:")
print("It includes 150 samples of Iris flowers, with 50 samples from each of three species: Iris setosa, Iris virginica, and Iris versicolor. There are four features measured in centimeters:")
print("- SepalLengthCm: Length of the sepal.")
print("- SepalWidthCm: Width of the sepal.")
print("- PetalLengthCm: Length of the petal.")
print("- PetalWidthCm: Width of the petal.")
print("The fifth column is 'Species', which is the target variable indicating the species of the Iris flower.")
print("\nSource:")
print("The dataset is provided locally as '/content/Iris.csv'. It is a widely used public dataset, originally published by Ronald Fisher in 1936, and is often sourced from the UCI Machine Learning Repository or similar public data repositories.")

print("\nFirst 5 rows of the dataset:")
print(df.head())

print("\nDataset Info:")
df.info()

print("\nDescriptive statistics of the dataset:")
print(df.describe())

# --- 2. Apply data manipulation techniques ---

print("\n--- Data Manipulation Techniques ---")

# Filtering specific rows based on conditions
print("\n1. Filtering: Rows where 'Species' is 'Iris-setosa'")
filtered_setosa = df[df['Species'] == 'Iris-setosa']
print(filtered_setosa.head())

print("\nFiltering: Rows where 'SepalLengthCm' > 5.5 and 'PetalLengthCm' < 4.0")
filtered_complex = df[(df['SepalLengthCm'] > 5.5) & (df['PetalLengthCm'] < 4.0)]
print(filtered_complex.head())

# Grouping data by categorical variables and computing aggregate statistics
print("\n2. Grouping and Aggregation: Aggregate statistics by 'Species'")
grouped_stats = df.groupby('Species').agg(
    count=('Id', 'count'), # Using Id column to count entries in each group
    mean_sepal_length=('SepalLengthCm', 'mean'),
    sum_petal_width=('PetalWidthCm', 'sum'),
    min_petal_length=('PetalLengthCm', 'min'),
    max_sepal_width=('SepalWidthCm', 'max')
)
print(grouped_stats)

print("\nMore detailed aggregation (mean, sum, min, max, count) for all numerical features by 'Species':")
detailed_grouped_stats = df.groupby('Species').agg(['mean', 'sum', 'min', 'max', 'count'])
print(detailed_grouped_stats)
```
