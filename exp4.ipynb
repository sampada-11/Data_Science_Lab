```python
def perform(data):
    if not data:
        return None # Handle empty data case

    n = len(data)

    # Mean calculation
    mean = sum(data) / n

    # Median calculation
    sorted_data = sorted(data)
    mid_idx = n // 2
    if n % 2 == 0:
        median = (sorted_data[mid_idx - 1] + sorted_data[mid_idx]) / 2
    else:
        median = sorted_data[mid_idx]

    # Variance calculation
    sqr_diff_sum = sum((val - mean) ** 2 for val in data)
    variance = sqr_diff_sum / (n - 1) # Use n for consistency

    # Standard deviation calculation
    std_deviation = variance ** 0.5

    # Return results as a dictionary
    return {
        "mean": mean,
        "median": median,
        "variance": variance,
        "std_deviation": std_deviation
    }

sample_data = [1,5,9,7,3,4,6,8,2]
stats = perform(sample_data)

if stats: # Check if stats were calculated (i.e., data was not empty)
    print(f"Mean: {stats['mean']:.2f}")
    print(f"Median: {stats['median']:.2f}")
    print(f"Variance: {stats['variance']:.2f}")
    print(f"Standard Deviation: {stats['std_deviation']:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate a normal distribution
# Parameters for the normal distribution
mu, sigma = 0, 0.1 # mean and standard deviation
s_normal = np.random.normal(mu, sigma, 1000) # 1000 samples

# 2. Plot its histogram
count, bins, ignored = plt.hist(s_normal, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

```python
import numpy as np

# Sample dataset
data = [1, 5, 8, 4, 9, 3, 2, 6, 7, 10]

# Compute statistics using numpy
mean = np.mean(data)
median = np.median(data)
variance = np.var(data, ddof=1) # ddof=1 for sample variance
std_deviation = np.std(data, ddof=1) # ddof=1 for sample standard deviation

print(f"Dataset: {data}")
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_deviation:.2f}")
```

```python
import numpy as np

# 4. Simulate a coin toss experiment and compute probabilities

# Number of coin tosses
num_tosses = 1000

# Simulate coin tosses (0 for tails, 1 for heads)
# np.random.randint(0, 2, num_tosses) generates random integers between 0 (inclusive) and 2 (exclusive)
coin_toss_results = np.random.randint(0, 2, num_tosses)

# Count heads and tails
heads_count = np.sum(coin_toss_results == 1)
tails_count = np.sum(coin_toss_results == 0)

# Compute probabilities
prob_heads = heads_count / num_tosses
prob_tails = tails_count / num_tosses

print(f"Coin Toss Simulation ({num_tosses} tosses):")
print(f"Heads: {heads_count} (Probability: {prob_heads:.2f})")
print(f"Tails: {tails_count} (Probability: {prob_tails:.2f})")
```
