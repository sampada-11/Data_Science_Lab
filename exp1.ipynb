**Factorial**

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Example usage:
while True:
    try:
        user_input = input("Enter a non-negative integer to calculate its factorial (or 'quit' to exit): ")
        if user_input.lower() == 'quit':
            break
        number = int(user_input)
        print(f"The factorial of {number} is {factorial(number)}")
    except ValueError:
        print("Invalid input. Please enter a non-negative integer or 'quit'.")
```

**Multiplication Table**

```python
#multiplication table
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):  # Table from 1 to 10
        print(f"{number} x {i} = {number * i}")

# Example usage:
while True:
    try:
        user_input = input("Enter an integer to display its multiplication table (or 'quit' to exit): ")
        if user_input.lower() == 'quit':
            break
        num = int(user_input)
        display_multiplication_table(num)
        print("\n") # Add a newline for better readability
    except ValueError:
        print("Invalid input. Please enter an integer or 'quit'.")
```

**Even Odd**

```python
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
while True:
    try:
        user_input = input("Enter an integer to check if it's odd or even (or 'quit' to exit): ")
        if user_input.lower() == 'quit':
            break
        number = int(user_input)
        print(check_odd_even(number))
    except ValueError:
        print("Invalid input. Please enter an integer or 'quit'.")
```

**Check Palindrome**

```python
#palindrome
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the original string with its reverse
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage:
while True:
    try:
        user_input = input("Enter an integer to check if it's a palindrome (or 'quit' to exit): ")
        if user_input.lower() == 'quit':
            break
        number = int(user_input)
        print(is_palindrome(number))
    except ValueError:
        print("Invalid input. Please enter an integer or 'quit'.")
```

# Task
Test the expert system's decision-making by inputting characteristics into the user interface and observing the inferred animal.

```python
# 1. Knowledge Base: Define the knowledge base for animal identification
knowledge_base = [
    {
        'conditions': ['has fur', 'barks', 'is domesticated'],
        'conclusion': 'Dog'
    },
    {
        'conditions': ['lives in water', 'has fins', 'has gills'],
        'conclusion': 'Fish'
    },
    {
        'conditions': ['has feathers', 'can fly', 'lays eggs'],
        'conclusion': 'Bird'
    },
    {
        'conditions': ['has rough skin', 'has four legs', 'lives in water and land'],
        'conclusion': 'Frog'
    },
    {
        'conditions': ['has fur', 'meows', 'is domesticated'],
        'conclusion': 'Cat'
    },
    {
        'conditions': ['has scales', 'crawls', 'no legs'],
        'conclusion': 'Snake'
    }
]

# 2. Inference Engine: Implement the inference engine
def infer_animal(characteristics):
    for rule in knowledge_base:
        # Check if all conditions in the rule are present in the provided characteristics
        if all(condition in characteristics for condition in rule['conditions']):
            return rule['conclusion']
    return 'Unknown animal based on provided characteristics.'

# 3. User Interface: Simple interactive user interface
print("Welcome to the Animal Identification Expert System!")
print("Enter characteristics as a comma-separated list (e.g., 'has fur, barks, is domesticated').")
print("Type 'quit' to exit.")

while True:
    user_input = input("\nEnter characteristics: ")
    if user_input.lower() == 'quit':
        print("Exiting expert system. Goodbye!")
        break

    # Process user input
    # Remove leading/trailing spaces and convert to lowercase for consistent matching
    # Also remove any surrounding quotes that users might include in their input
    input_characteristics = [char.strip().lower().replace("'", "").replace('"', '') for char in user_input.split(',')]

    # Infer the animal
    inferred_animal = infer_animal(input_characteristics)
    print(f"Inferred animal: {inferred_animal}")
```
