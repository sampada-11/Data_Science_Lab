**1. WAP to compute factorial of a given number**

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Example usage:
while True:
    try:
        user_input = input("Enter a non-negative integer to calculate its factorial (or 'quit' to exit): ")
        if user_input.lower() == 'quit':
            break
        number = int(user_input)
        print(f"The factorial of {number} is {factorial(number)}")
    except ValueError:
        print("Invalid input. Please enter a non-negative integer or 'quit'.")
```

**2. WAP to display table for a given number**

```python
#multiplication table
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):  # Table from 1 to 10
        print(f"{number} x {i} = {number * i}")

# Example usage:
while True:
    try:
        user_input = input("Enter an integer to display its multiplication table (or 'quit' to exit): ")
        if user_input.lower() == 'quit':
            break
        num = int(user_input)
        display_multiplication_table(num)
        print("\n") # Add a newline for better readability
    except ValueError:
        print("Invalid input. Please enter an integer or 'quit'.")
```

**3. WAP to identify the given number is odd or even**

```python
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
while True:
    try:
        user_input = input("Enter an integer to check if it's odd or even (or 'quit' to exit): ")
        if user_input.lower() == 'quit':
            break
        number = int(user_input)
        print(check_odd_even(number))
    except ValueError:
        print("Invalid input. Please enter an integer or 'quit'.")
```

**4. WAP to identify the entered number is palindrome or not.**

```python
#palindrome
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the original string with its reverse
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage:
while True:
    try:
        user_input = input("Enter an integer to check if it's a palindrome (or 'quit' to exit): ")
        if user_input.lower() == 'quit':
            break
        number = int(user_input)
        print(is_palindrome(number))
    except ValueError:
        print("Invalid input. Please enter an integer or 'quit'.")
```

**5. Develop an Expert System that provides simple decision-making.**
**Basic Loan Approval Expert System**


```python
import re

# 1. Knowledge Base: Define the knowledge base for the loan approval expert system
loan_knowledge_base = [
    {
        'conditions': ['credit_score >= 700', 'annual_income >= 50000', 'employed', 'debt_to_income_ratio < 0.30'],
        'conclusion': 'Approved'
    },
    {
        'conditions': ['credit_score >= 600', 'annual_income >= 30000', 'employed', 'debt_to_income_ratio < 0.40'],
        'conclusion': 'Approved with conditions'
    },
    {
        'conditions': ['credit_score < 600'],
        'conclusion': 'Denied (Low Credit Score)'
    },
    {
        'conditions': ['annual_income < 20000', 'not employed'],
        'conclusion': 'Denied (Low Income/Unemployed)'
    },
    {
        'conditions': ['debt_to_income_ratio >= 0.50'],
        'conclusion': 'Denied (High Debt-to-Income Ratio)'
    },
    {
        'conditions': ['credit_score >= 650', 'annual_income >= 40000', 'employed', 'debt_to_income_ratio >= 0.30', 'debt_to_income_ratio < 0.50'],
        'conclusion': 'Refer for manual review'
    }
]

# 2. Inference Engine: Implement the inference engine
def infer_loan_decision(applicant_characteristics, knowledge_base):
    """
    Infers a loan decision based on applicant characteristics and a knowledge base.

    Args:
        applicant_characteristics (dict): A dictionary containing the applicant's attributes
                                        (e.g., 'credit_score', 'annual_income', 'employed',
                                        'debt_to_income_ratio').
        knowledge_base (list): A list of rule dictionaries, where each rule has 'conditions'
                               and a 'conclusion'.

    Returns:
        str: The inferred loan decision (e.g., 'Approved', 'Denied', 'Refer for manual review',
             or 'No conclusive decision').
    """
    for rule in knowledge_base:
        all_conditions_met = True
        for condition_str in rule['conditions']:
            try:
                # Evaluate the condition string within a safe scope.
                # The '__builtins__': {} prevents access to Python's built-in functions,
                # and applicant_characteristics provides the local variables for evaluation.
                result = eval(condition_str, {"__builtins__": {}}, applicant_characteristics)
                if not result:
                    all_conditions_met = False
                    break  # Break from inner loop if a condition is not met
            except (NameError, TypeError, SyntaxError) as e:
                # If a characteristic is missing, condition is malformed, or any eval error occurs,
                # consider the condition not met for this rule.
                # print(f"Warning: Could not evaluate condition '{condition_str}' due to error: {e}. Assuming condition not met.")
                all_conditions_met = False
                break

        if all_conditions_met:
            return rule['conclusion']

    return 'No conclusive decision'

# 3. User Interface: Create an interactive user interface
print("Welcome to the Loan Approval Expert System!")
print("Please provide the following applicant characteristics (type 'quit' to exit):")

while True:
    try:
        credit_score_input = input("Enter credit score (e.g., 720): ")
        if credit_score_input.lower() == 'quit':
            break
        credit_score = int(credit_score_input)

        annual_income_input = input("Enter annual income (e.g., 60000): ")
        if annual_income_input.lower() == 'quit':
            break
        annual_income = float(annual_income_input)

        employed_input = input("Is the applicant employed? (yes/no): ")
        if employed_input.lower() == 'quit':
            break
        employed = employed_input.lower() == 'yes'

        debt_to_income_ratio_input = input("Enter debt-to-income ratio (e.g., 0.25): ")
        if debt_to_income_ratio_input.lower() == 'quit':
            break
        debt_to_income_ratio = float(debt_to_income_ratio_input)

        applicant_data = {
            'credit_score': credit_score,
            'annual_income': annual_income,
            'employed': employed,
            'not employed': not employed, # To handle 'not employed' condition in rules
            'debt_to_income_ratio': debt_to_income_ratio
        }

        decision = infer_loan_decision(applicant_data, loan_knowledge_base)
        print(f"Loan Decision: {decision}\n")

    except ValueError:
        print("Invalid input. Please ensure numeric values are entered correctly and 'yes'/'no' for employment.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

print("Exiting Loan Approval Expert System. Goodbye!")
```
