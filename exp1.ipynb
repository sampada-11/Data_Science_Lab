**Multiplication Table**

```python
#multiplication table
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):  # Table from 1 to 10
        print(f"{number} x {i} = {number * i}")

# Example usage:
num = 7
display_multiplication_table(num)

print("\n") # Add a newline for better readability between examples

num = 12
display_multiplication_table(num)
```

**Check Palindrome**

```python
#palindrome
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the original string with its reverse
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage:
num1 = 121
print(is_palindrome(num1))

num2 = 123
print(is_palindrome(num2))

num3 = 1001
print(is_palindrome(num3))
```

**Even Odd**

```python
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
num1 = 10
print(check_odd_even(num1))

num2 = 7
print(check_odd_even(num2))

num3 = 0
print(check_odd_even(num3))
```

**Factorial**

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Example usage:
number = 5
print(f"The factorial of {number} is {factorial(number)}")

number = 0
print(f"The factorial of {number} is {factorial(number)}")

number = -3
print(f"The factorial of {number} is {factorial(number)}")
```

### Summary of Expert System Capabilities

**1. Knowledge Base Structure:**
The expert system's knowledge base is structured as a Python list of dictionaries. Each dictionary represents a rule for identifying an animal. A rule consists of two key components:
*   `'conditions'`: A list of strings, where each string is a characteristic that must be present for the rule to be true (e.g., `['has fur', 'barks', 'is domesticated']`).
*   `'conclusion'`: A string representing the animal identified if all conditions are met (e.g., `'Dog'`)

This structure allows for a clear and organized representation of domain-specific knowledge, in this case, for animal identification.

**2. Inference Mechanism:**
The inference engine is implemented as a Python function named `infer_animal`. It operates by taking a list of user-provided characteristics and iterating through each rule in the knowledge base. For each rule, it checks if *all* conditions specified in the rule's `'conditions'` list are present within the user's characteristics. The first rule whose conditions are fully satisfied leads to its `'conclusion'` being returned as the inferred animal. If no rules in the knowledge base have all their conditions met by the user's input, the engine concludes that the animal is "Unknown animal based on provided characteristics."

**3. User Interface (UI):**
The system features a basic text-based user interface that allows for interactive querying. Users are prompted to enter a comma-separated list of animal characteristics. The UI processes this input, converts it into a list of individual characteristics, and then passes it to the `infer_animal` function. The inferred animal (or the "Unknown" message) is then displayed back to the user. The UI also includes a loop for continuous queries and a 'quit' command to gracefully exit the system, enhancing user experience.

**4. Domain of the Expert System:**
The expert system's domain is **animal identification**. It is designed to classify animals based on a set of predefined characteristics stored in its knowledge base.

**5. Overall Capabilities and Limitations:**
*   **Capabilities:** The system effectively identifies animals for which clear rules exist in its knowledge base and for which sufficient characteristics are provided by the user. It accurately matched `Dog`, `Fish`, and `Frog` in testing when the exact conditions were met. Its forward-chaining inference mechanism is straightforward and efficient for this type of rule-based system.
*   **Limitations:** The system is entirely dependent on the completeness and accuracy of its knowledge base. It cannot handle ambiguous inputs, partial matches (unless designed to do so, which it currently is not), or characteristics that are not explicitly part of any rule. As observed in testing, if a characteristic is misspelled or a combination of characteristics doesn't perfectly align with a defined rule, it defaults to "Unknown animal based on provided characteristics," even if some characteristics might hint at a known animal. This highlights its strictly rule-following nature and lack of fuzzy logic or learning capabilities.
