```python

#Create a 3x3 square matrix and compute its determinant.(WITHOUT NUMPY)

def determinant_3x3(matrix):
    """
    Calculate determinant of 3x3 matrix using the formula:
    det = a(ei − fh) − b(di − fg) + c(dh − eg)
    """
    a = matrix[0][0]
    b = matrix[0][1]
    c = matrix[0][2]
    d = matrix[1][0]
    e = matrix[1][1]
    f = matrix[1][2]
    g = matrix[2][0]
    h = matrix[2][1]
    i = matrix[2][2]

    det = (a * (e * i - f * h) -
           b * (d * i - f * g) +
           c * (d * h - e * g))

    return det

# Take user input for 3x3 matrix
print("Enter elements of 3x3 matrix row by row:")
matrix = []

for i in range(3):
    row = []
    print(f"Enter row {i+1} (3 numbers separated by space):")
    row_input = input().strip().split()

    # Convert strings to floats
    for num in row_input:
        row.append(float(num))

    matrix.append(row)

# Display the matrix
print("\nYour 3x3 matrix:")
for row in matrix:
    print(row)

# Calculate and display determinant
det = determinant_3x3(matrix)
print(f"\nDeterminant = {det}")
```

```python
#Find the inverse of the matrix.(WITHOUT NUMPY)
def determinant_3x3(matrix):
    """Calculate determinant of 3x3 matrix"""
    a, b, c = matrix[0]
    d, e, f = matrix[1]
    g, h, i = matrix[2]

    return (a*(e*i - f*h) - b*(d*i - f*g) + c*(d*h - e*g))

def adjoint_3x3(matrix):
    """Calculate adjoint (adjugate) matrix"""
    a, b, c = matrix[0]
    d, e, f = matrix[1]
    g, h, i = matrix[2]

    # Cofactor matrix
    cofactor = [
        [e*i - f*h, -(d*i - f*g), d*h - e*g],
        [-(b*i - c*h), a*i - c*g, -(a*h - b*g)],
        [b*f - c*e, -(a*f - c*d), a*e - b*d]
    ]

    # Transpose to get adjoint
    adjoint = [
        [cofactor[0][0], cofactor[1][0], cofactor[2][0]],
        [cofactor[0][1], cofactor[1][1], cofactor[2][1]],
        [cofactor[0][2], cofactor[1][2], cofactor[2][2]]
    ]

    return adjoint

def inverse_3x3(matrix):
    """Calculate inverse = (1/det) * adjoint"""
    det = determinant_3x3(matrix)

    if abs(det) < 1e-10:  # Check if singular matrix
        return None

    adj = adjoint_3x3(matrix)

    # Multiply adjoint by 1/determinant
    inv = []
    for i in range(3):
        row = []
        for j in range(3):
            row.append(adj[i][j] / det)
        inv.append(row)

    return inv

# Take user input for 3x3 matrix
print("Enter elements of 3x3 matrix row by row:")
matrix = []

for i in range(3):
    print(f"Enter row {i+1} (3 numbers separated by space):")
    row = list(map(int, input().strip().split())) # Changed float to int
    matrix.append(row)

print("\nYour matrix:")
for row in matrix:
    print(f"  {row[0]:6d} {row[1]:6d} {row[2]:6d}") # Changed float formatting to int

# Calculate inverse
inverse_matrix = inverse_3x3(matrix)

if inverse_matrix is None:
    print("\n Matrix is singular (determinant = 0). No inverse exists!")
else:
    print("\n Inverse matrix:")
    for row in inverse_matrix:
        print(f"  {row[0]:6.2f} {row[1]:6.2f} {row[2]:6.2f}")
```

```python
#Solve the following system of linear equations: 2x+3y=10, 4x+5y=20(WITHOUT NUMPY)
def solve_2x2_system(a1, b1, c1, a2, b2, c2):
    """
    Solve: a1*x + b1*y = c1
           a2*x + b2*y = c2
    """
    # Using Cramer's Rule (super simple!)
    det = a1*b2 - a2*b1

    if abs(det) < 1e-10:
        return None  # No unique solution

    x = (c1*b2 - c2*b1) / det
    y = (a1*c2 - a2*c1) / det

    return x, y

# Take user input
print("Enter coefficients for 2 equations (2x + 3y = 10, 4x + 5y = 20):")
print("Equation 1: a1 b1 c1 (for a1*x + b1*y = c1)")
a1, b1, c1 = map(float, input("Enter a1 b1 c1: ").strip().split())

print("Equation 2: a2 b2 c2 (for a2*x + b2*y = c2)")
a2, b2, c2 = map(float, input("Enter a2 b2 c2: ").strip().split())

# Solve
solution = solve_2x2_system(a1, b1, c1, a2, b2, c2)

if solution is None:
    print("\n No unique solution (determinant = 0)")
else:
    x, y = solution
    print(f"\n Solution:")
    print(f"x = {x:.2f}")
    print(f"y = {y:.2f}")

    # Verify
    print(f"\nVerification:")
    print(f"{a1:.1f}x + {b1:.1f}y = {a1*x + b1*y:.1f} (should be {c1})")
    print(f"{a2:.1f}x + {b2:.1f}y = {a2*x + b2*y:.1f} (should be {c2})")
```

```python
import numpy as np
#Solve the following system of linear equations using NumPy: 2x+3y=10, 4x+5y=20


print("Enter coefficients for the system of linear equations:")
print("For equation 1: a1*x + b1*y = c1")
a1 = float(input("Enter a1: "))
b1 = float(input("Enter b1: "))
c1 = float(input("Enter c1: "))

print("\nFor equation 2: a2*x + b2*y = c2")
a2 = float(input("Enter a2: "))
b2 = float(input("Enter b2: "))
c2 = float(input("Enter c2: "))

# Define the coefficient matrix A
A = np.array([[a1, b1], [a2, b2]])

# Define the constant vector b
b = np.array([c1, c2])

print("\nCoefficient Matrix A:")
print(A)
print("\nConstant Vector b:")
print(b)

# Solve the system of equations
try:
    solution_np = np.linalg.solve(A, b)
    x_np, y_np = solution_np
    print(f"\nSolution using NumPy:")
    print(f"x = {x_np:.2f}")
    print(f"y = {y_np:.2f}")

    # Verification
    print(f"\nVerification:")
    print(f"{a1:.1f}x + {b1:.1f}y = {a1*x_np + b1*y_np:.1f} (should be {c1})")
    print(f"{a2:.1f}x + {b2:.1f}y = {a2*x_np + b2*y_np:.1f} (should be {c2})")
except np.linalg.LinAlgError:
    print("\nError: The matrix is singular or ill-conditioned. No unique solution exists.")
```

```python
# Calculate the inverse using NumPy
import numpy as np

# Function to take matrix input from user
def get_matrix_input(rows, cols):
    matrix = []
    print(f"Enter elements of the {rows}x{cols} matrix row by row:")
    for i in range(rows):
        while True:
            try:
                row_input = input(f"Enter row {i + 1} (space-separated numbers): ").strip().split()
                if len(row_input) != cols:
                    raise ValueError(f"Please enter exactly {cols} numbers for row {i + 1}.")
                row = [float(x) for x in row_input]
                matrix.append(row)
                break
            except ValueError as e:
                print(f"Invalid input: {e}. Please try again.")
    return np.array(matrix)

# Get 3x3 matrix input from the user
user_matrix = get_matrix_input(3, 3)

print("\nYour 3x3 matrix:")
print(user_matrix)

# Calculate the inverse using NumPy
try:
    numpy_inverse_matrix = np.linalg.inv(user_matrix)
    print("\nInverse matrix (NumPy):")
    print(numpy_inverse_matrix)

    # Verification (optional: multiply matrix by its inverse to get identity matrix)
    print("\nVerification (Matrix * Inverse - Identity Matrix, should be close to zero):")
    identity_check = np.dot(user_matrix, numpy_inverse_matrix)
    print(identity_check - np.identity(3))

except np.linalg.LinAlgError:
    print("\nError: The matrix is singular, so its inverse cannot be computed.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
